```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

plt.style.use("classic")
%matplotlib inline
```

```python
def constraint(x):
    return 10 - x  

def indiff(x):
    return -5*x + 50
```

```python
x = np.linspace(0, 100, 100)
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1,1,1)
ax.plot(constraint(x), x, c="black", lw=2)
ax.plot(x, indiff(x), "--", c="black")
ax.plot(10, 0, "o", markersize=15, c="orange")
ax.set_title("$Optimal\ Games$", fontsize=25, y=1.02)
ax.set_ylim([0, 12])
ax.set_xlim([0, 12])
ax.set_xlabel("$Games\ of\ League\ of\ Legends$", fontsize=15)
ax.set_ylabel("$Games\ of\ DotA$", fontsize=15)
for tick in ax.get_xticklabels():
    tick.set_fontname("Georgia")
for tick in ax.get_yticklabels():
    tick.set_fontname("Georgia")
ax.text(1, 9.5, "$Time$\n$Constraint$", fontsize=15)
ax.text(8.25, 10.25, "$Indifference$\n$Curve$", fontsize=15)
ax.text(10, 0.5, "($League^{*}, DotA^{*})$\n$(10, 0)$", fontsize=10)
```
